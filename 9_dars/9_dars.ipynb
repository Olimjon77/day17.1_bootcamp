```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("customer_purchase_data")
```

```python
df.head()
```

```python
df["Internet usage"].value_counts()
```

```python
sns.set_style("whitegrid")
sns.catplot(x = "Internet usage",
            y = "Age",
            data = df,
            hue = "Gender",
            kind = "point")
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.catplot(x = "Internet usage",
            y = "Age",
            data = df,
            hue = "Gender",
            kind = "point")
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.set_palette("RdBu")
sns.set_style("darkgrid")
sns.catplot(x = "Internet usage",
            data = df,
            hue = "Gender",
            kind = "count")
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.set_context("talk")
sns.set_palette("RdBu")
sns.set_style("darkgrid")
sns.catplot(x = "Internet usage",
            data = df,
            hue = "Gender",
            kind = "count")
plt.xticks(rotation = 30)
plt.show()
```

```python
sns.set_context("notebook")
sns.set_palette("RdBu")
sns.set_style("dark")
sns.catplot(x = "Internet usage",
            data = df,
            hue = "Gender",
            kind = "count")
plt.xticks(rotation = 30)
plt.grid()
plt.show()
```

```python
g  =sns.catplot(x ="Internet usage",
                y = "Age",
                data =  df,
                kind = "box")
g.figure.suptitle("New Title")
plt.xticks(rotation= 30)
plt.show()
```

```python
g  =sns.catplot(x ="Internet usage",
                y = "Age",
                data =  df,
                kind = "box",
                col = "Gender")
g.figure.suptitle("New Title", y=1.2)
plt.xticks(rotation= 30)
plt.show()
```

```python
g  =sns.catplot(x ="Internet usage",
                y = "Age",
                data =  df,
                kind = "box",
                col = "Gender")
g.figure.suptitle("New Title", y=1.2)
g.set_titles("This is {col_name} table")
plt.xticks(rotation= 30)
plt.show()
```

```python
url ="https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv"
df2 = pd.read_csv(url)
df2.head()
```

```python
sns.displot(df2["Tuition"])
sns.despine(left= True)
```

```python
sns.set_theme(color_codes=True)
sns.displot(df2["Tuition"], color= "g")
sns.despine(left= True)
```

```python
fig, ax = plt.subplots()
sns.histplot(df2['Tuition'], ax= ax)
ax.set(xlabel = "Tuition 2013-14",
       ylabel = "Distribution", xlim=(0,50000),
       title= "2013-14 Tuition and Feels Distribution")
```

```python
sns.set_style("darkgrid")
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2,
                               sharey=True, figsize= (7,4))
sns.histplot(df2["Tuition"], stat = "density", ax =ax0)
sns.histplot(df2.query('REGION== 5')["Tuition"], stat  = 'density', ax = ax1)
ax1.set(xlabel= "Tuition(MN)", xlim=(0,70000))
ax1.axvline(x= 30000, label= "My Budget", linestyle = "--")
ax1.legend()
```
