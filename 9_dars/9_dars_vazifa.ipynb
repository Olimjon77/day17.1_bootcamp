```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_purchase_data.csv")
```

```python

df= df.iloc[:100]
```

```python
df.head()
```

```python
df.loc[df["Gender"]==0, "Gender"] = "Male"
df.loc[df["Gender"]==1, "Gender"] = "Female"
df.head()
```

```python
df.loc[df["ProductCategory"]==0, "ProductCategory"] = "Electronics"
df.loc[df["ProductCategory"]==1, "ProductCategory"] = "Clothing"
df.loc[df["ProductCategory"]==2, "ProductCategory"] = "Home Goods"
df.loc[df["ProductCategory"]==3, "ProductCategory"] = "Beauty"
df.loc[df["ProductCategory"]==4, "ProductCategory"] = "Sports"
df.head(10)
```

```python
sns.set_style("whitegrid")
sns.catplot(x= "ProductCategory",
             y= "Age",
             data = df,
             hue = "Gender",
             kind= "point")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.catplot(y= "ProductCategory",
             x= "Age",
             data = df,
             hue = "Gender",
             kind= "point")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set_style("ticks")
sns.catplot(x= "ProductCategory",
             y= "Age",
             data = df,
             hue = "Gender",
             kind= "point")
plt.xticks(rotation = 90)
plt.show()
```

```python
catgory_order= [1,2,3,4]
sns.catplot(x= "ProductCategory",
            data=df,
            kind = "count",
            order =catgory_order )
plt.show()
```

```python
sns.set_palette("GnBu")
catgory_order= [1,2,3,4]
sns.catplot(x= "ProductCategory",
            data=df,
            kind = "count",
            order =catgory_order )
plt.show()
```

```python
custom_palette= ["r", "b", "m", "y"]
sns.set_palette(custom_palette)
catgory_order= [1,2,3,4]
sns.catplot(x= "ProductCategory",
            data=df,
            kind = "count",
            order =catgory_order )
plt.show()
```

```python
sns.set_context("talk")
custom_palette= ["r", "b", "m", "y"]
sns.set_palette(custom_palette)
catgory_order= [1,2,3,4]
sns.catplot(x= "ProductCategory",
            data=df,
            kind = "count",
            order =catgory_order )
plt.show()
```

```python
sns.set_context("notebook")
custom_palette= ["r", "b", "m", "y"]
sns.set_palette(custom_palette)
catgory_order= [1,2,3,4]
sns.catplot(x= "ProductCategory",
            data=df,
            kind = "count",
            order =catgory_order )
plt.show()
```

```python
g= sns.catplot(x= "PurchaseStatus",
               y= "DiscountsAvailed",
                data=df,
                    kind = "box",
)
g.fig.suptitle("New Title")
plt.show()
```
