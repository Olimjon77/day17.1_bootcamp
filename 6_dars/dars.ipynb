```python
medals = pd.read_csv('medals_by_country_2016.csv', index_col = 0)
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
plt.show()
```

```python
medals['Gold'] + medals['Silver']
```

```python

fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom= medals['Gold'])
ax.bar(medals.index, medals['Bronze'], bottom= medals['Gold'] + medals['Silver'])
ax.set_xticklabels(medals.index, rotation=45)
plt.show()
```

```python

table = pd.read_csv('summer2016.csv', index_col=0)
table.head(2)
```

```python
mens_rowing = table[(table['Sex'] == 'M') & (table['Sport'] == 'Rowing')]

mens_gymnastics = table[(table['Sex'] == 'M') & (table['Sport'] == 'Gymnastics')]

fix, ax = plt.subplots()
ax.bar("Rowing", mens_rowing["Height"].mean())  
ax.bar("Gymnastics", mens_gymnastics["Height"].mean())  
ax.set_ylabel("Height (cm)") 
plt.show() 

```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label='Rowing')
ax.hist(mens_gymnastics['Height'], label='Gymnastics')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label='Rowing', bins=5)
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=5)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing['Height'], label='Rowing', bins=5, histtype='step')
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=5, histtype='step')
ax.legend()
plt.show()
```
