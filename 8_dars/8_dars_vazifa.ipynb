```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("penguins.csv")
```

```python
df.head()
```

```python
sns.scatterplot( x= df["species"], 
                y = df['body_mass_g'] )
```

```python
sns.countplot(x = df["island"])
plt.show()
```

```python
sns.countplot(x = "species",
              data = df)
```

## Seaborn kutubxonasidan scatter plot orqali diagramma chizish

```python
sns.scatterplot(x= "species",
                y = "body_mass_g", data = df)
```

## Hue bilan rang orqali farqlash

```python
sns.scatterplot(x= "species",
                y = "body_mass_g", 
                data = df,
                hue= "sex")
```

## Hue_colors va palette orqali farqlash

```python
hue_colors = {"MALE": "k",
              "FEMALE": "r",
              "NaN" : "m"}
sns.scatterplot(x= "species",
                y = "body_mass_g", 
                data = df,
                hue= "sex",
                palette= hue_colors )
plt.show()
```

```python

hue_colors = {"MALE": "k",
              "FEMALE": "r",
              "NaN" : "m"}
sns.countplot(x= "island",
            
            data = df,
            hue= "sex",
            palette= hue_colors)
plt.show()
```

## Kind orali relplotda scatter ni tanlash

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter")
plt.show()
```

## Rowe orqali alohida diagramma chiqarish

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter",
            row = "sex")
plt.show()
```

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter",
            row = "island")
plt.show()
```

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter",
            row = "species")
plt.show()
```

## Col va row orqali har birini alohida ko'rish 

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter",
            row = "sex",
            col= "species")
plt.show()
```

```python
sns.relplot(x= "body_mass_g",
            y= "flipper_length_mm",
            data= df,
            kind = "scatter",
            row = "sex",
            col= "species",
            col_order= ["Adelie", "Chinstrap"])
plt.show()
```

## Size orqali katta kichik qlib farqlash

```python
sns.relplot(x= "bill_depth_mm",
            y= "bill_length_mm",
            data= df,
            kind= "scatter",
            size= "body_mass_g")
plt.show()
```

```python
sns.relplot(x= "bill_depth_mm",
            y= "bill_length_mm",
            data= df,
            kind= "scatter",
            size= "body_mass_g",
            hue= "body_mass_g")
plt.show()
```

## Alpha orqli shaffofligini boshqarish

```python
sns.relplot(x= "bill_depth_mm",
            y= "bill_length_mm",
            data= df,
            kind= "scatter",
            size= "body_mass_g",
            hue= "body_mass_g",
            alpha= 0.7)
plt.show()
```

```python
sns.relplot(x= "bill_depth_mm",
            y= "bill_length_mm",
            data= df,
            kind= "line",
            size= "body_mass_g",
            hue= "body_mass_g")
plt.show()
```
