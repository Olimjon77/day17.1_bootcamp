```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
df= df.drop(columns='species')
```

```python
df.head()
```

```python
df["sepal_width"].hist()
```

```python
sns.histplot(x= "sepal_width",
             data= df,
             kde= True)
```

```python
df["sepal_width"].mean()
```

```python
df["sepal_width"].describe()
```

```python
df["new_sepal_width"]= (df["sepal_width"] - df["sepal_width"].mean())/df["sepal_width"].std()
```

```python
df
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
fig.set_size_inches(12, 8)
sns.histplot(df["new_sepal_width"],  kde = True, ax =ax0)
sns.histplot(df["sepal_width"], kde = True, ax =ax1)
```

```python
new_df = (df - df.mean())/df.std()
```

```python
new_df
```

```python
df
```

```python
new_df = (df-df.min())/(df.max()- df.min())
```

```python
new_df
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
fig.set_size_inches(12, 8)
sns.histplot(df["sepal_width"],  kde = True, ax =ax0)
sns.histplot(new_df["sepal_width"], kde = True, ax =ax1)
```

```python
new1_df = (df-df.mean())/(df.max()- df.min())
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
fig.set_size_inches(10, 4)
sns.histplot(df["sepal_width"],  kde = True, ax =ax0)
sns.histplot(new1_df["sepal_width"], kde = True, ax =ax1)
```

```python
df.head()
```

```python
df["maxabsscaling"]= df["sepal_width"] / df["sepal_width"].abs().max()
```

```python
df
```

```python
df["robust_scalling_s_w"] = (df["sepal_width"] - df["sepal_width"].median())/(df["sepal_width"].quantile(0.75)-df["sepal_width"].quantile(0.25))
```

```python
df
```

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
fig.set_size_inches(10, 4)
sns.histplot(df["sepal_width"],  kde = True, ax =ax0)
sns.histplot(df["robust_scalling_s_w"], kde = True, ax =ax1)
```
