## Obyektni numerikka o'tkazish

```python
df["COMPL_RPY_5YR_RT"] = pd.to_numeric(df["COMPL_RPY_5YR_RT"], errors="coerce")
df["DEBT_MDN"] = pd.to_numeric(df["DEBT_MDN"], errors="coerce")
df["MEDIAN_HH_INC"] = pd.to_numeric(df["MEDIAN_HH_INC"], errors="coerce")
```

```python
df.dtypes
```

```python
# df.dropna().reset_index(drop=True)   
```

```python
df["COMPL_RPY_5YR_RT"].isna().sum()
```

## Transform orqali nanlarni ozgartirish

```python
df["COMPL_RPY_5YR_RT_FILLNA_median"] = df.groupby(["REGION", "Ownership"])["COMPL_RPY_5YR_RT"].transform(lambda x: x.fillna(x.median()))
df["COMPL_RPY_5YR_RT_FILLNA_mean"] = df.groupby(["REGION", "Ownership"])["COMPL_RPY_5YR_RT"].transform(lambda x: x.fillna(x.mean()))

```

##  Plotta hist orqali bir nechta malumotlarni bitta ifodalash

```python
fig, ax = plt.subplots(figsize= (10,6))
df["COMPL_RPY_5YR_RT"].plot(kind="hist", ax=ax, alpha=0.5, color="blue", bins=20, label= "Numeric")
df["COMPL_RPY_5YR_RT_FILLNA_median"].plot(kind="hist", ax=ax, alpha= 0.5, color= "green", bins= 20, label= "Fillna_Median")
df["COMPL_RPY_5YR_RT_FILLNA_mean"].plot(kind="hist", ax=ax, alpha= 0.5, color= "red", bins= 20, label= "Fillna_Mean")
ax.set_ylabel("Value")
ax.set_ylabel("Frequency")
ax.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize= (12,8))
df["COMPL_RPY_5YR_RT"].plot(kind="kde", ax=ax, alpha=0.5, color="blue",label= "Numeric")
df["COMPL_RPY_5YR_RT_FILLNA_median"].plot(kind="kde", ax=ax, alpha= 0.5, color= "green", label= "Fillna_Median")
df["COMPL_RPY_5YR_RT_FILLNA_mean"].plot(kind="kde", ax=ax, alpha= 0.5, color= "red", label= "Fillna_Mean")
ax.set_ylabel("Value")
ax.set_ylabel("Frequency")
ax.legend()
ax.set_title("COMPL_RPY_5YR_RT")
plt.grid()
plt.show()
```

```python
df["DEBT_MDN_FILLNA_median"] = df.groupby(["REGION", "Ownership"])["DEBT_MDN"].transform(lambda x: x.fillna(x.median()))
df["DEBT_MDN_FILLNA_mean"] = df.groupby(["REGION", "Ownership"])["DEBT_MDN"].transform(lambda x: x.fillna(x.mean()))

```
