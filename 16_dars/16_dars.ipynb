```python
df["COMPL_RPY_5YR_RT_NUMERIC"]=pd.to_numeric(df["COMPL_RPY_5YR_RT"], errors="coerce")
df.groupby(by=["Regions", "Ownership"])["COMPL_RPY_5YR_RT_NUMERIC"].agg(['count', 'mean', 'median'])
df["COMPL_RPY_5YR_RT_NUMERIC_median"] = df.groupby(['REGION', 'Ownership'])['COMPL_RPY_5YR_RT_NUMERIC'].transform(lambda x: x.fillna(x.median()))
```
