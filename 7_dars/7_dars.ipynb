```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather = pd.read_csv('seattle_weather.csv')
seattle_weather.head(15)
```

```python
station1 = seattle_weather[seattle_weather['STATION'] == 'USC00456295']['MLY-TAVG-NORMAL']
station2 = seattle_weather[seattle_weather['STATION'] == 'USW00024222']['MLY-TAVG-NORMAL']
```

```python
fig, ax = plt.subplots(2,1)
ax[0].scatter(seattle_weather['MLY-PRCP-NORMAL'], seattle_weather['MLY-TAVG-NORMAL'])
ax[0].set_ylabel('Temperature (C)')

ax[1].bar('USC00456295', station1.mean(), yerr=station1.std())
ax[1].bar('USW00024222', station2.mean(), yerr=station2.std())
plt.show()
```
