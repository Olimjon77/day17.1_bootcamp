```python
url = "https://assets.datacamp.com/production/repositories/2210/datasets/fb4f2c1039e3df2c2e2624a8c95de5a1980861c6/bike_share.csv"
df2= pd.read_csv(url)
df2.head()
```

```python
sns.regplot(data=df2, x="temp",
            y= "total_rentals",
             marker='+' )
```

```python
sns.residplot(data =df2,
              x= "temp", y= "total_rentals")
```

```python
sns.regplot(data =df2,
              x= "temp", y= "total_rentals",
              order=2)
```

```python
sns.regplot(data= df2,
            x= "mnth",
            y= "total_rentals",
            x_jitter=.1, order= 2)
```

```python
import numpy as np
```

```python
sns.regplot(data= df2,
            x= "mnth",
            y= "total_rentals",
            x_estimator=np.mean, order= 2)
```

```python
sns.regplot(data= df2,
            x= "temp",
            y= "total_rentals",
            order=2,
            x_bins=4)
```

```python
sns.heatmap(pd.crosstab(df2["mnth"], df2["weekday"],
            values=df2["total_rentals"], aggfunc="mean"))
```

```python
sns.heatmap(pd.crosstab(df2["mnth"], df2["weekday"],
            values=df2["total_rentals"], aggfunc="mean").astype(int), annot = True, fmt="d",
            cmap="YlGnBu", cbar=True, linewidths=0.5)
```

```python
sns.heatmap(pd.crosstab(df2["mnth"], df2["weekday"],
            values=df2["total_rentals"], aggfunc="mean").astype(int), annot = True, fmt="d",
            cmap="YlGnBu", cbar=True, center=pd.crosstab(df2["mnth"], df2["weekday"],
            values=df2["total_rentals"], aggfunc="mean").loc[9,3] )
```
