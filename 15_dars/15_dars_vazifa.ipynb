```python
df["Dep_Time"] = pd.to_timedelta(df["Dep_Time"]+":00")
```

```python
df["Arrival_Time"] = df["Date_of_Journey"]+ df["Duration"]+ df["Dep_Time"]
```

```python
df
```

```python
df.dtypes
```

```python
df["Total_Stops"].value_counts()
```

```python
stop_mapping={'non-stop':0,
              '1 stop':1,
              '2 stops':2,
              '3 stops':3,
              '4 stops':4
}
df['Total_Stops_mapped'] = df['Total_Stops'].map(stop_mapping)
```

```python
df["Total_Stops_mapped"].value_counts()
```

```python
df["month"] = df["Date_of_Journey"].dt.month
df["weekday"] = df["Date_of_Journey"].dt.weekday
df
```

```python
twenty_fifth = df["Price"].quantile(0.25)
median = df["Price"].median()
seventy_fifth = df["Price"].quantile(0.75)
maximum = df["Price"].max()
```

```python
labels = ["Economy", "Premium Economy", "Business Class", "First Class"]  
bins = [0, twenty_fifth, median, seventy_fifth, maximum]
```

```python
df["Price_Category"] = pd.cut(df["Price"],     
                            labels=labels,                              
                             bins=bins)
```
